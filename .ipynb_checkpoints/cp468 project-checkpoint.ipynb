```python
import cv2
import tensorflow as tf

def maker(filepath):
    
    sizeofimg = 50  # 50 in txt-based
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (sizeofimg, sizeofimg))
    return new_array.reshape(-1, sizeofimg, sizeofimg, 1)


model = tf.keras.models.load_model("64x3-CNN.model")
def prediction(model):
    guess = model.predict([maker('bishop1.jpg')])
    print(guess)  # will be a list in a list.
    count=0
    for i in guess[0]:
        if(i==1.0):
            break
        else:
         count+=1
    if(count==0):
        print("bishop")
    elif(count==1):    
        print("knight")
    elif(count==2):
        print("pawn")
    elif(count==3):
        print("queen")
    elif(count==4):
        print("rook")
prediction(model)        
        
    
```
